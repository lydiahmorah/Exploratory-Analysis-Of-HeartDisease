# **18.Plot the continious variables**

```python
cate_val = []
cont_val = []
for column in data.columns:
    if len(data[column].unique())<=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cont_val
```

```python
#plotting the continous variables add title Key Health Indicators in Relation to Heart Disease
plt.figure(figsize=(15,6))
plt.suptitle('Key Health Indicators in Relation to Heart Disease')
data[cont_val].hist(figsize=(15,6),color='blue')
plt.tight_layout()
plt.show()
```

```python
#replace 3 in the thal column woth the mode of the column
data.thal.replace(3,data.thal.mode()[0],inplace=True)
```

```python
data.thal.unique()
```

```python
#plotting the  barplot thal and rename xaxis as Blood disorder and rename unique types as
# 0 as Normal,1 as Fixed Defect and 2 as Reversible Defect

#plotting thal
sns.barplot(x='thal',y='age',data=data)
plt.xlabel('Blood Disorder')
plt.ylabel('Age')
plt.show()
```

```python
#getting the count for each thal category
data.thal.value_counts()

# renaming the unique types in thal column 0 as Normal,1 as Fixed Defect and 2 as Reversible Defect
data.thal.replace({0:'Normal',1:'Fixed Defect',2:'Reversible Defect'},inplace=True)
data.thal.value_counts()
```

```python
#bivariate analysis comparing thal and the target variable
sns.countplot(x='thal',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.xlabel('Blood Disorder')
plt.ylabel('Count')
plt.show()
```
